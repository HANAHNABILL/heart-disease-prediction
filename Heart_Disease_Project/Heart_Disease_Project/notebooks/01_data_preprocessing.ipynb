```python
# Heart Disease UCI - Data Preprocessing

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.impute import SimpleImputer
import warnings
warnings.filterwarnings('ignore')


sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

# Load the dataset WITHOUT headers (since it doesn't have column names)
df = pd.read_csv('../data/heart_disease.csv', header=None)

# Add proper column names based on the UCI Heart Disease dataset description
column_names = [
    'age',      # age in years
    'sex',      # sex (1 = male; 0 = female)
    'cp',       # chest pain type (1: typical angina, 2: atypical angina, 3: non-anginal pain, 4: asymptomatic)
    'trestbps', # resting blood pressure (in mm Hg)
    'chol',     # serum cholesterol in mg/dl
    'fbs',      # fasting blood sugar > 120 mg/dl (1 = true; 0 = false)
    'restecg',  # resting electrocardiographic results (0: normal, 1: ST-T wave abnormality, 2: left ventricular hypertrophy)
    'thalach',  # maximum heart rate achieved
    'exang',    # exercise induced angina (1 = yes; 0 = no)
    'oldpeak',  # ST depression induced by exercise relative to rest
    'slope',    # slope of the peak exercise ST segment (1: upsloping, 2: flat, 3: downsloping)
    'ca',       # number of major vessels (0-3) colored by fluoroscopy
    'thal',     # thalassemia (3 = normal; 6 = fixed defect; 7 = reversible defect)
    'target'    # target variable (0 = no disease, 1-4 = disease presence levels)
]

df.columns = column_names

# Display basic information
print("Dataset Shape:", df.shape)
print("\nDataset Columns:", df.columns.tolist())
print("\nFirst 10 rows:")
display(df.head(10))

print("\nDataset Info:")
df.info()

print("\nSummary Statistics:")
display(df.describe())

# Check for missing values (represented as '?' in the data)
print("\nChecking for missing values (any '?' in the data):")
for col in df.columns:
    # Check if there are any non-numeric values
    unique_vals = df[col].unique()
    print(f"{col}: {[val for val in unique_vals if not isinstance(val, (int, float)) or pd.isna(val)]}")

# Convert all columns to numeric
for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

print("\nMissing Values after conversion:")
print(df.isnull().sum())

# Handle missing values
if df.isnull().sum().sum() > 0:
    print("\nImputing missing values...")
    # use mean imputation
    numerical_cols = df.select_dtypes(include=[np.number]).columns
    imputer = SimpleImputer(strategy='mean')
    df[numerical_cols] = imputer.fit_transform(df[numerical_cols])

# Convert target to binary classification (0 = no disease, 1 = any disease)
# 0 = no disease, 1-4 = different levels of disease
df['target_binary'] = (df['target'] > 0).astype(int)

print(f"\nOriginal target distribution:")
print(df['target'].value_counts().sort_index())
print(f"\nBinary target distribution:")
print(df['target_binary'].value_counts())
print(f"\nBinary target proportions:")
print(df['target_binary'].value_counts(normalize=True))

# Separate features and target (using binary target for classification)
X = df.drop(['target', 'target_binary'], axis=1)
y = df['target_binary']  # Using binary classification

print(f"\nFeature matrix shape: {X.shape}")
print(f"Target vector shape: {y.shape}")


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Convert back to DataFrame for easier handling
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

print("\nFeatures after scaling:")
display(X_scaled.head())

# Exploratory Data Analysis

# 1. Distribution of target variable
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# Original target distribution
df['target'].value_counts().sort_index().plot(kind='bar', ax=ax1, color='skyblue')
ax1.set_title('Original Target Distribution\n(0=No Disease, 1-4=Disease Levels)')
ax1.set_xlabel('Target Value')
ax1.set_ylabel('Count')

# Binary target distribution
df['target_binary'].value_counts().plot(kind='bar', ax=ax2, color='lightcoral')
ax2.set_title('Binary Target Distribution\n(0=No Disease, 1=Disease)')
ax2.set_xlabel('Target Value')
ax2.set_ylabel('Count')
plt.tight_layout()
plt.show()

# 2. Correlation heatmap
plt.figure(figsize=(12, 10))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, fmt='.2f')
plt.title('Feature Correlation Heatmap')
plt.tight_layout()
plt.show()

# 3. Boxplots for numerical features
numerical_features = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.ravel()  # Flatten the axes array

for i, col in enumerate(numerical_features):
    sns.boxplot(x=y, y=df[col], ax=axes[i])
    axes[i].set_title(f'{col} by Heart Disease')
    
# Remove empty subplot
for i in range(len(numerical_features), 6):
    fig.delaxes(axes[i])
    
plt.tight_layout()
plt.show()

# 4. Histograms of features
df[numerical_features].hist(figsize=(15, 10), bins=20)
plt.tight_layout()
plt.show()

# Save the preprocessed data
preprocessed_data = X_scaled.copy()
preprocessed_data['target'] = y.values
preprocessed_data['target_original'] = df['target'].values

preprocessed_data.to_csv('../data/heart_disease_processed.csv', index=False)
print("\nPreprocessed data saved successfully to '../data/heart_disease_processed.csv'")

# Also save the cleaned original data with binary target
df.to_csv('../data/heart_disease_cleaned.csv', index=False)
print("Cleaned data saved successfully to '../data/heart_disease_cleaned.csv'")

# Save the scaler for later use
import joblib
joblib.dump(scaler, '../models/scaler.pkl')
print("Scaler saved to '../models/scaler.pkl'")

print("\nData preprocessing completed successfully!")
```

```python
import pandas as pd
import numpy as np

# Load your dataset first
df = pd.read_csv('../data/heart_disease_cleaned.csv')  # Use the cleaned/original data, not scaled

print("Dataset shape:", df.shape)
print("\nFirst 5 rows:")
print(df.head())

# Age-adjusted heart rate function
def adjust_heart_rate(age, thalach):
    max_predicted_hr = 220 - age  # Standard clinical formula
    return thalach / max_predicted_hr  # Percentage of predicted max

# Apply the adjustment
df['thalach_relative'] = df.apply(lambda x: adjust_heart_rate(x['age'], x['thalach']), axis=1)

print("\n" + "="*60)
print("AGE-ADJUSTED HEART RATE ANALYSIS")
print("="*60)

print("\nOriginal vs Age-Adjusted Heart Rate:")
print(df[['age', 'thalach', 'thalach_relative']].head(10))

print("\nYoung patients (age < 40) analysis:")
young_patients = df[df['age'] < 40]
print(f"Number of young patients: {len(young_patients)}")
if len(young_patients) > 0:
    print(young_patients[['age', 'thalach', 'thalach_relative', 'target']])

print("\n" + "="*60)
print("YOUR PROBLEMATIC TEST CASE ANALYSIS")
print("="*60)

# Simulate your 25-year-old female case
test_age = 25
test_thalach = 119
test_relative_hr = adjust_heart_rate(test_age, test_thalach)

print(f"Test case - Age: {test_age}, Max HR: {test_thalach}")
print(f"Age-predicted max HR: {220 - test_age}")
print(f"Relative HR (% of predicted max): {test_relative_hr:.1%}")

# Compare with actual young patients in dataset
similar_cases = df[(df['age'] >= 20) & (df['age'] <= 30)]
print(f"\nSimilar aged patients in dataset (20-30 years): {len(similar_cases)}")
if len(similar_cases) > 0:
    print(similar_cases[['age', 'thalach', 'thalach_relative', 'target']])

print("\n" + "="*60)
print("CLINICAL INTERPRETATION")
print("="*60)

if test_relative_hr < 0.6:
    print("🚨 ACHTUNG: 60% of predicted max HR indicates SEVERE exercise intolerance")
    print("This would be clinically concerning for a 25-year-old")
elif test_relative_hr < 0.8:
    print("⚠️  80% of predicted max HR is below average for young adults")
    print("Might indicate moderate exercise limitation")
else:
    print("✅ Normal exercise capacity for age")

# Check what heart rate would be normal
normal_hr_25 = (220 - 25) * 0.85  # 85% of predicted max = good effort
print(f"\nExpected max HR for good effort (25yo): {normal_hr_25:.0f} bpm")
print(f"Your input (119 bpm) is {((test_thalach/normal_hr_25)-1):.1%} below expected")
```
